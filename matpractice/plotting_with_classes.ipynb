We will be using Python to make a random walk and plot it in a visually appealing manner.

A random walk is a path that is determined by a series of decisions left to chance

```python
from random import choice
import matplotlib.pyplot as plt
```

```python
class RandomWalk:
    """A class to generate random walks"""

    def __init__(self, num_points=5000):
        """Initialize attributes of a walk"""

        self.num_points = num_points # If not value is provided 5000 as default

        # All walks start at (0, 0)
        self.x_values = [0]
        self.y_values = [0]

    def fill_walk(self):
        """Calculate all the points in the walk."""
        # Keep taking steps until the walk reaches the desired length
        while len(self.x_values) < self.num_points:

            # Decide which direction to go, and how far.
            x_direction = choice([1, -1]) # Right or Left
            x_distance = choice([0, 1, 2, 3, 4])
            x_step = x_direction * x_distance

            y_direction = choice([1, -1]) # Up or Down
            y_distance = choice([0, 1, 2, 3, 4])
            y_step = y_direction * y_distance

            # Reject moves that go nowhere
            if x_step == 0 and y_step == 0:
                continue

            # Calculate new position
            x = self.x_values[-1] + x_step # We add the last value in x_values to the new step
            y = self.y_values[-1] + y_step # We add the last value in y_values to the new step

            self.x_values.append(x) # Append to end of list
            self.y_values.append(y)
```

```python
rw = RandomWalk()

rw.fill_walk()

# Plot main points of the walk
plt.style.use('classic')
fig, ax = plt.subplots()
ax.scatter(rw.x_values, rw.y_values, s=10)
# ax.set_aspect('equal') normally you would use this but its kinda ugly in jupyter notebook, it makes sure the spacing between tick marks is equal on the x and the y axes
plt.show()
```

```python
rw = RandomWalk()

rw.fill_walk()

# Plot main points of the walk
plt.style.use('classic')
fig, ax = plt.subplots()
point_numbers = range(rw.num_points)

# This config emphasizes where the walk started and where it ended
ax.scatter(
    rw.x_values, rw.y_values, c=point_numbers, cmap=plt.cm.Blues, edgecolors='none', s=15
)

# ax.set_aspect('equal') normally you would use this but its kinda ugly in jupyter notebook, it makes sure the spacing between tick marks is equal on the x and the y axes
plt.show()
```
