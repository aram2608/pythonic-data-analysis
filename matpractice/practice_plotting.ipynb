````python
import matplotlib.pyplot as plt
````

````python
sqaures = [1, 4, 9, 16, 25]
````

A simple line graph using Matplotlib

````python
 # the variable ax represents a single plot in the figure, while fig represents all the figures in the plot
fig, ax = plt.subplots() # can generate one or more plots in the same figure
ax.plot(sqaures)
plt.show()
````

Not bad, but we can make it a lot prettier and more informative

````python
fig, ax = plt.subplots()
ax.plot(sqaures, linewidth=3) # alters line width

# Setting chart and axes titles
ax.set_title("Square Numbers", fontsize=24)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=14)

# Set size of tick labels
ax.tick_params(labelsize=14)

plt.show()
````

Much better yeah? Matplotlib can take two sets of values however, in the previous plot we only used one.

````python
input_values = [1, 2, 3, 4, 5]
sqaures = [1, 4, 9, 16, 25]

fig, ax = plt.subplots()
ax.plot(input_values, sqaures, linewidth=3) # inputs go in this order example.plot(x_values, y_values)

# Setting chart and axes titles
ax.set_title("Square Numbers", fontsize=24)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=14)

# Set size of tick labels
ax.tick_params(labelsize=14)
````

````python
input_values = [1, 2, 3, 4, 5]
sqaures = [1, 4, 9, 16, 25]

# set style before you make the plot
plt.style.use('seaborn-v0_8')
fig, ax = plt.subplots()
ax.plot(input_values, sqaures, linewidth=3) # inputs go in this order example.plot(x_values, y_values)

# Setting chart and axes titles
ax.set_title("Square Numbers", fontsize=24)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=14)

# Set size of tick labels
ax.tick_params(labelsize=14)
````

Calculating data automatically.

We can use loops to do calculations for us.

````python
x_values = range(1, 1000)
y_values = [x**2 for x in x_values]
# this is list comprehension
# this is the same as
# y_values = []
# for x in x_values:
# new_x = x**2
# y_values.append(new_x)

plt.style.use('seaborn-v0_8')
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, s=10)
ax.set_title("Square Numbers", fontsize=24)
ax.set_xlabel("Values", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=14)

# Setting axes range
ax.axis([0, 1100, 0, 1_100_000]) # the first two are minimum and maximum of the x axis, the second two are for the y axis
ax.ticklabel_format(style='plain')

plt.show()
````

````python
x_values = range(1, 1000)
y_values = [x**2 for x in x_values]

plt.style.use('seaborn-v0_8')
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, color='red',s=10) # we can use the color parameter to define a set of colors, RBG values can be used
ax.set_title("Square Numbers", fontsize=24)
ax.set_xlabel("Values", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=14)

# Setting axes range
ax.axis([0, 1100, 0, 1_100_000]) # the first two are minimum and maximum of the x axis, the second two are for the y axis
ax.ticklabel_format(style='plain')

plt.show()
````

````python
x_values = range(1, 1000)
y_values = [x**2 for x in x_values]

plt.style.use('seaborn-v0_8')
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, c=y_values, cmap=plt.cm.Blues,s=10) # we can use a colormap to show changes in values
ax.set_title("Square Numbers", fontsize=24)
ax.set_xlabel("Values", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=14)

# Setting axes range
ax.axis([0, 1100, 0, 1_100_000]) # the first two are minimum and maximum of the x axis, the second two are for the y axis
ax.ticklabel_format(style='plain')

plt.show()
````
